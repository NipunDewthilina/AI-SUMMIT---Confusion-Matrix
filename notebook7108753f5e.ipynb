```python
dusage_avg = [i for i in range(0,7500,100)]
train['dusage_avg_bin']=pd.cut(train['dusage_avg'],dusage_avg,labels=dusage_avg[1:],retbins=False)
```

```python
dusage_sum = [i for i in range(0,178000,1000)]
train['dusage_sum_bin']=pd.cut(train['dusage_sum'],dusage_sum,labels=dusage_sum[1:],retbins=False)
dusage_min = [i for i in range(0,3700,100)]
train['dusage_min_bin']=pd.cut(train['dusage_min'],dusage_min,labels=dusage_min[1:],retbins=False)
dusage_max = [i for i in range(0,30700,100)]
train['dusage_max_bin']=pd.cut(train['dusage_max'],dusage_max,labels=dusage_max[1:],retbins=False)
```

```python
train['dusage_avg_bin'] = train['dusage_avg_bin'].astype('float64').replace(np.nan, 0)
train['dusage_sum_bin'] = train['dusage_sum_bin'].astype('float64').replace(np.nan, 0)
train['dusage_min_bin'] = train['dusage_min_bin'].astype('float64').replace(np.nan, 0)
train['dusage_max_bin'] = train['dusage_max_bin'].astype('float64').replace(np.nan, 0)
```

```python
vusage_onnet_avg = [i for i in range(0,240,10)]
train['vusage_onnet_avg_bin']=pd.cut(train['vusage_onnet_avg'],vusage_onnet_avg,labels=vusage_onnet_avg[1:],retbins=False)
#vusage_onnet_sum = [i for i in range(0,178000,1000)]
#train['vusage_onnet_sum_bin']=pd.cut(train['vusage_onnet_sum'],vusage_onnet_sum,labels=vusage_onnet_sum[1:],retbins=False)
#vusage_onnet_min = [i for i in range(0,3700,100)]
#train['vusage_onnet_min_bin']=pd.cut(train['vusage_onnet_min'],vusage_onnet_min,labels=vusage_onnet_min[1:],retbins=False)
#vusage_onnet_max = [i for i in range(300,30700,100)]
#train['vusage_onnet_max_bin']=pd.cut(train['vusage_onnet_max'],vusage_onnet_max,labels=vusage_onnet_max[1:],retbins=False)
```

```python
vusage_offnet_avg = [i for i in range(0,120,10)]
train['vusage_offnet_avg_bin']=pd.cut(train['vusage_offnet_avg'],vusage_offnet_avg,labels=vusage_offnet_avg[1:],retbins=False)
vusage_offnet_sum = [i for i in range(0,2660,10)]
train['vusage_offnet_sum_bin']=pd.cut(train['vusage_offnet_sum'],vusage_offnet_sum,labels=vusage_offnet_sum[1:],retbins=False)
vusage_offnet_min = [i for i in range(0,40,10)]
train['vusage_offnet_min_bin']=pd.cut(train['vusage_offnet_min'],vusage_offnet_min,labels=vusage_offnet_min[1:],retbins=False)
vusage_offnet_max = [i for i in range(0,490,10)]
train['vusage_offnet_max_bin']=pd.cut(train['vusage_offnet_max'],vusage_offnet_max,labels=vusage_offnet_max[1:],retbins=False)
```

```python
train['vusage_onnet_avg_bin'] = train['vusage_onnet_avg_bin'].astype('float64').fillna(0)
train['vusage_offnet_avg_bin'] = train['vusage_offnet_avg_bin'].astype('float64').replace(np.nan, 0)
train['vusage_offnet_sum_bin'] = train['vusage_offnet_sum_bin'].astype('float64').replace(np.nan, 0)
train['vusage_offnet_min_bin'] = train['vusage_offnet_min_bin'].astype('float64').replace(np.nan, 0)
train['vusage_offnet_max_bin'] = train['vusage_offnet_max_bin'].astype('float64').replace(np.nan, 0)
```

```python
train = pd.concat([train,pd.get_dummies(train.device_type,prefix='device_type_')],axis=1)
train = pd.concat([train,pd.get_dummies(train.device_category,prefix='device_category_')],axis=1)
train = pd.concat([train,pd.get_dummies(train.district_name,prefix='district_name_')],axis=1)
train = pd.concat([train,pd.get_dummies(train.age_group,prefix='age_group_')],axis=1)
```

```python
S = {'FEMALE':1,'MALE':0}
train = train.replace({"gender": S})
P = {'PKG1':1,"PKG2":2,'PKG3':3,'PKG4':4,'PKG5':5,'PKG6':6,'PKG7':7,'PKG8':8}
train = train.replace({"next_month_plan": P})
```
